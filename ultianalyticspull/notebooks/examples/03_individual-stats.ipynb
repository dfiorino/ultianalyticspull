```python
import pandas as pd
import glob
audl = pd.concat([pd.read_csv(dfteam, index_col=0) for dfteam in glob.glob('../../data/processed/*/*.csv')])
```

```python
goal_scores = audl[(audl.Action=='Goal')].groupby(['Tournament','Receiver','Teamname']).apply(lambda x : len(x))
goal_scores = goal_scores.reset_index().rename(columns={0:'Goals'}).sort_values(['Goals'],ascending=False)
goal_scores[:15].reset_index(drop=True)
```

```python
goal_asst = audl[(audl.Action=='Goal')].groupby(['Tournament','Passer','Teamname']).apply(lambda x : len(x))
goal_asst = goal_asst.reset_index().rename(columns={0:'Goals'}).sort_values(['Goals'],ascending=False)
goal_asst[:15].reset_index(drop=True)
```

```python
blks = audl[(audl.Action=='D')].groupby(['Tournament','Defender','Teamname']).apply(lambda x : len(x))
blks = blks.reset_index().rename(columns={0:'Goals'}).sort_values(['Goals'],ascending=False)
blks[:15].reset_index(drop=True)
```
