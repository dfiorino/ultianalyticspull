```python
import pandas as pd
import glob
import difflib
import numpy as np
```

```python
username_playername_relation = pd.read_csv('05_username_playername_relation.csv',encoding = "ISO-8859-1")
```

```python
def GetSuggestion1(row):
    return GetSuggestion(row,1)
def GetSuggestion2(row):
    return GetSuggestion(row,2)   
def GetSuggestion3(row):
    return GetSuggestion(row,3)  
def GetSuggestion(row,no):
    roster = username_playername_relation[username_playername_relation.Teamname==row['Teamname']].PlayerName.dropna().unique()
    matches = difflib.get_close_matches(row['Username'], roster)
    return matches[no-1] if len(matches) > no-1 else np.nan

username_playername_relation.loc[pd.isnull(username_playername_relation.PlayerName),
                                 'Suggestion1'] = username_playername_relation.loc[pd.isnull(username_playername_relation.PlayerName)].apply(GetSuggestion1,axis=1)
username_playername_relation.loc[pd.isnull(username_playername_relation.PlayerName),
                                 'Suggestion2'] = username_playername_relation.loc[pd.isnull(username_playername_relation.PlayerName)].apply(GetSuggestion2,axis=1)
username_playername_relation.loc[pd.isnull(username_playername_relation.PlayerName),
                                 'Suggestion3'] = username_playername_relation.loc[pd.isnull(username_playername_relation.PlayerName)].apply(GetSuggestion3,axis=1)
                                                                                                                                      
                                                                                                                                                        
                                                                                                                                                   
                                                                                                                                      
```
